```python
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups

docs = fetch_20newsgroups(subset='all',  remove=('headers', 'footers', 'quotes'))['data']

print(docs[0])
```

```python
import os

os.environ["TOKENIZERS_PARALLELISM"] = "false"

topic_model = BERTopic()
topics, probs = topic_model.fit_transform(docs)
print(topic_model.get_topic_info())
```

```python
topic_model.visualize_topics()
```

So here it is shown that the Yelp dataset is so diverse, that you get a wide variety of topics that are not very useful to specific reviews. Since the Yelp dataset is very large and does not seem to be sorted, the Amazon dataset will now be taken as they can be filtered on a specific category

```python
import pandas as pd

amazon_df = pd.read_csv('./amazon_sorted/most_populair_products.csv')
amazon_df = amazon_df.drop(columns='index')

topic_model = BERTopic(nr_topics=10)
topic_model.fit_transform(amazon_df['review_body'])
topic_model.get_topic_info()
```

```python
topic_model.visualize_topics()
```

```python
product_df = amazon_df[amazon_df['product_id'] == amazon_df.iloc[0]['product_id']]
topic_model.fit_transform(product_df['review_body'])
topic_model.get_topic_info()
```

```python
topic_model.visualize_topics()
```
